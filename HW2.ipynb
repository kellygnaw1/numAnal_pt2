#### use approximation to correct trapezoid rule: 
Now we can use our second order approximation for $f'(b)$ and $f'(a)$ to correct the Trapezoid rule. The third equation, $I(f;h) = I + \frac{h^2}{12}(f'(b) - f'(a)) + O(h^4)$, gives us the composite Trapezoid Rule approximation and the error so we can use this idea to correct the composite trapezoid rule, equation (1) by adding the error back. (Technically it depends on the error, the approximation may be under or over the actual so if it's under we need to add back the difference and if it's over we need to subtract it).  <br /><br />
$$I(f) = \frac{h}{2}f(x_0)+h\sum_{i=1}^{n-1}f(x_i)+\frac{h}{2}f(x_n)-\frac{h^2}{12}(\frac{3f(b)-4f(b-h) +f(b-2h)}{2h} - \frac{-3f(a)+4f(a+h) -f(a+2h)}{2h})+O(h^4)$$ <br />
This would then give us fourth degree accuracy because our error is fourth order. In the cell below, I will use $f(x) = e^x$ to verify that this is true numerically. 


```python
#Verify numerically for question 1
#we can see that our alpha values are around 4 so we can conclude that the quadrature formula has 4th degree accuracy
import numpy as np
from tabulate import tabulate
import matplotlib.pyplot as plt

def f(x):
    return np.exp(x)

def trapezoidal_rule(f, a, b, n):
    h = (b-a)/n
    x = np.arange(n+1)*h
    b_prime = (3*f(b) - 4*f(b-h) + f(b-2*h))/(2*h)
    a_prime = (-3*f(a) + 4*f(a+h) - f(a+2*h))/(2*h)
    result = (0.5*h*f(x[0]) + h*sum(f(x[1:n])) + 0.5*h*f(x[n])) - (((h**2)/12)*((b_prime - a_prime)))
    return h, result 

a = 0.
b = 1.

NSamples = 10 #Number of sample values of h (or n)
dx = [0]*NSamples
quad = [0]*NSamples      
Integral = np.exp(1.) - 1.
err = [0]*NSamples      
alpha = [0]*NSamples
alpha[0] = 0

n = 1 # Initial value of n

for i in range(NSamples):
    n = n * 2
    [dx[i], quad[i]] = trapezoidal_rule(f,a,b,n) 
    err[i] = abs(quad[i] - Integral)
    if i > 0:
        alpha[i] = np.log(err[i]/err[i-1])/np.log(dx[i]/dx[i-1])

results = [(dx[i], quad[i], err[i], alpha[i]) for i in range(NSamples)]
print(tabulate(results, headers=["h", "T(f,a,b,n)", "|T(f,a,b,n) - I[f;a,b]|", "alpha"], floatfmt=".15g",tablefmt="presto"))

fig, ax = plt.subplots(figsize=(9,6))
ax.loglog(dx,err,'bo-',label='Trapezoidal',lw=2)
ax.loglog(dx,dx,'ro-',label='err(h) = h',lw=2)
ax.set_title("Error in Trapezoid Rule approximation",fontsize=22)
ax.legend(fontsize=15)
ax.set_xlabel('h',fontsize=22)
ax.set_ylabel('err(h)',fontsize=22)
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
```

We know that a quadratic formula comes in the form of $f(x) = ax^2 + bx + c$  for some constants $a, b, c$. This means we can rewrite the integral as $\int_{0}^{1}ax^2+bx+c dx = \left(\frac{ax^3}{3}\right)\bigg|^{0}_{1} + \left(\frac{bx^2}{2}\right)\bigg|^{0}_{1} + \left(cx\right)\bigg|^{0}_{1} = \frac{a}{3}+\frac{b}{2} + c$. Now suppose $\int_{0}^{1}f(x) = \alpha[f(x_0) + f(x_1)] = \frac{a}{3}+\frac{b}{2} + c$ then we have $\frac{a}{3}+\frac{b}{2} + c = \alpha[ax_0^{2} + bx_0 + c + ax_1^{2} +bx_1 +c] = a\alpha(x_0^{2}+x_1^{2}) + b\alpha(x_0+x_1) + 2c\alpha$. We can then find our constants by first evaluating c to find the alpha value. <br /><br />
$$c = 2c\alpha, \alpha = \frac{1}{2}$$
$$\frac{a}{3} = a\alpha(x_0^{2}+x_1^{2}) = \frac{a}{2}(x_0^{2}+x_1^{2}) \implies \frac{2}{3} = x_0^{2}+x_1^{2}$$
$$\frac{b}{2} = b\alpha(x_0+x_1) = \frac{b}{2}(x_0+x_1) \implies 1 = x_0 + x_1$$
Now we can write $2(x_0^2+x_1^2) = (x_0+x_1)^2+(x_0-x_1)^2$
$$1+(x_0-x_1)^2=\frac{4}{3} \implies (x_0-x_1)^2 = \frac{1}{3} \implies (x_0-x_1) = \frac{1}{\sqrt{3}} $$
From this we can find $x_0, x_1$ since $x_0+x_1 = 1$ and $x_0-x_1 =  \frac{1}{\sqrt{3}}$. ## Exercise 4
Suppose $\int_{-1}^{1}f(x)dx \approx \sum_{i=0}^{n}A_if(x_i)$ is exact for all polynomial of degree $\leq n$ where $n$ is even and all nodes are symmetrically placed around the origin. We want to show that the formula is exact for all polynomials of degree $n+1$. <br /><br />

Let $P$ be a polynomial where $P(x) = \pi_{i=0}^n(x-x_i)$ for $i = 0,\dots,n$. Now let $P'$ be a polynomial of degree n+1. Since both $P(x)$ and $P'(x)$ are polynomials of degree n+1, any polynomial $f(x) = cP(x) + p(x)$ will be an element in a polynomial set of degree n+1 for c is some nonzero constant and $p(x)$ is some polynomial of degree $\leq n$. This means for $j \in{0, \dots, n}$ there exists one term in $P(x_j)$ that will be of the form $(x_j-x_j)$ so our entire $P(x_j)$ will equal to 0. <br/><br/>

Then for some constant $A_j$ we have $A_jf(x_j) = A_j[cP(x_j) + p(x_j)] = A_jp(x_j)$. This tells us that $I(f) = I(p)$ because $I(f) = \sum_{i=0}^{n}A_if(x_i) = \sum_{i=0}^{n}A_ip(x_i)=I(p)$. Since we are assuming that the nodes are symmetric around the origin and $f$ has degree $n+1$, the formula is therefore exact for polynomials with degree $n+1$.
