Suppose our functions have sufficient derivatives. We know that the error for approximation an integral using Simpson's rule for $n=2$ is $\frac{-1}{90}(\frac{b-a}{2})^5f(\xi)$ where $\xi$ is some number between the interval $a, b$. This is for the simpson rule hence we only have $b-a$, if we were to use the Composite Simpson's rule then we can replace $b-a$ with smaller, more precise intervals. Let $h$ represent the smaller intervals in Composite Simpson's rule, to then find the error, we need to add up the errors of all the subintervals. <br/><br/>

Suppose we have $n$ subintervals, then the error we get must be less than $\frac{-n}{90}(\frac{h}{2})^5||f^4||_{\infty}$ since this denotes for the maximum error. By dividing the initial interval into subintervals, we know the error we get will be less than or equal to the original so if we let $h = b-a$ we have $\frac{-1}{90}(\frac{h}{2})^4||f||_{\infty}$. Thus the Composite Simpson's rule is $O(h^4)$.
